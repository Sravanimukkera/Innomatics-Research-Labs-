```python
#Task-4
#1.Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False

    return True


while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount):
            print("Recharge Successful!")
            break
        else:
            print("Please try again.\n")

    except ValueError:
        print("Please enter a valid number.\n")
```

```python
#2.Inventory Reorder Alert System

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 5
}

check_inventory(inventory)
```

```python
#3.Student Result Processing System
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


marks_list = [60, 45, 70, 50]

average, result = process_result(marks_list)

print("Average Marks:", average)
print("Result:", result)
```

```python
#4.Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print("Total Fare: ₹", total_fare)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

    except ValueError:
        print("Please enter valid input.\n")
```

```python
#5.Employee Attendance Eligibility Checker

def check_eligibility(attendance):
    total_days = len(attendance)
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

percentage, status = check_eligibility(attendance_list)

print("Attendance Percentage:", percentage)
print("Status:", status)
```

```python
#6.Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
result = check_password(pwd)

print(result)
```
