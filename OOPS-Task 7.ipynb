```python
#Task -7 OOPS

#1.Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

light_name = input()
action = input()

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
#2.Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

name = input()
emp_id = input()
department = input()

employee = Employee(name, emp_id, department)
employee.display_id_card()
```

```python
#3.Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input()
phone = input()

c = Contact(name, phone)
c.display()
```

```python
#4.Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

name = input()
price = input()

p = Product(name, price)
p.display()
```

```python
#5.Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", str(self.rating), "/ 5")

name = input()
rating = input()

m = Movie(name, rating)
m.display()
```

```python
#6.Delivery Address Manager

class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)

customer = input()
address = input()

d = Delivery(customer, address)
d.display()
```
