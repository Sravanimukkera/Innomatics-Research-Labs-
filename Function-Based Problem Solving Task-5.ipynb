```python
#Task-5 Real-World Function-Based Problem Solving
#1.Social Media – Post Engagement Analyzer

def post_engagement_analyzer(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

likes = [200, 150, 300, 400]
post_engagement_analyzer(likes)

```

```python
#2.Healthcare – Medicine Stock Alert System

def medicine_stock_alert(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

medicine_stock_alert(6)
```

```python
#3.Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_list, required_level):
    total = 0
    count = 0

    for rain in rainfall_list:
        total += rain
        count += 1

    average = total // count

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_data = [70, 75, 68, 80, 67]
rainfall_checker(rainfall_data, 70)
```

```python
#4.Duplicate Account Detection

def duplicate_account_detection(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user2", "user3", "user1"]
duplicate_account_detection(users)
```

```python
#5.Appointment Eligibility Checker

def appointment_eligibility(age):
    print("Patient Age:", age)

    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

appointment_eligibility(21)
```

```python
#6.Premium Crop Price Filter

def premium_crop_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

crop_prices = [1500, 2500, 1800, 3200]
premium_crop_filter(crop_prices)
```

```python
#7.System Monitoring – Application Health Checker

def application_health_checker(error_count):
    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

application_health_checker(7)
```

```python
#8.Banking – Daily Transaction Limit Checker

def transaction_limit_checker(amount):
    print("Transaction Amount:", amount)

    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

transaction_limit_checker(60000)
```

```python
#9.E-Learning – Student Attendance Eligibility System

def attendance_eligibility(attendance_list):
    total = len(attendance_list)
    present = 0

    for status in attendance_list:
        if status == "P":
            present += 1

    percentage = (present / total) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance = ["P", "P", "A", "P", "P", "P", "P", "P"]
attendance_eligibility(attendance)
```

```python
#10.Smart Electricity Bill Analyzer

def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

units_consumed = 180
total_bill, usage_status = electricity_bill(units_consumed)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
