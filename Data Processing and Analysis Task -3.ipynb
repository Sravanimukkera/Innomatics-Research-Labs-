```python
#Task-3
#1.Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
#2.Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.replace('.', '').replace(',', '').replace('!', '').replace('?', '')
words = query.split()

word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

frequent_words = {}
for word, count in word_count.items():
    if count > 1:
        frequent_words[word] = count

print(frequent_words)


```

```python
#3.Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4.Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
total = len(emails)

for email in emails:
    domain = email.split('@')[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#5.Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)

for day, value in enumerate(sales, start=1):
    if value > average * 1.3:
        print(f"Day {day}: {value}")
```

```python
#6.Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}
for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1
    else:
        id_count[uid] = 1

for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
