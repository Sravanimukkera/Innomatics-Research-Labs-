```python
#Function-Based Problem Solving-Task 6
#1.Smart Parking Lot Management System

def smart_parking_system(capacity, vehicle_logs):
    currently_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            currently_parked += 1
        elif log == "OUT":
            if currently_parked > 0:
                currently_parked -= 1

        if currently_parked > peak_usage:
            peak_usage = currently_parked

    if currently_parked > capacity:
        status = "Over Capacity - Alert!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", currently_parked)
    print("Parking Status:", status)
    print("Peak Parking Usage:", peak_usage)


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
#2.Online Food Delivery Time Estimator

def delivery_time_estimator(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "Low":
        traffic_delay = 0
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 15
    else:
        weather_delay = 0

    total_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", total_time, "minutes")


distance = 8
traffic = "High"
weather = "Rainy"

delivery_time_estimator(distance, traffic, weather)
```

```python
#3.Movie Theatre Seat Occupancy Analyzer

def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", str(int(occupancy)) + "%")
    print("Show Status:", status)


total_seats = 200
booked_seats = [1] * 150

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
#4.Cloud Server Load Classification System

def server_load_classification(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif 50 <= average < 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", str(int(average)) + "%")
    print("Server Status:", status)


cpu_readings = [45, 60, 70, 85, 90]

server_load_classification(cpu_readings)
```

```python
#5.Smart Classroom Resource Usage Monitor

def resource_usage_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if len(overused) > 0:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", energy_alert)


resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_usage_monitor(resource_usage)
```

```python
#6.Online Event Registration Capacity Controller

def event_capacity_controller(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


capacity = 100
registrations = 105

event_capacity_controller(capacity, registrations)
```
